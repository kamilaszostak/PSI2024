Stwórz i zaimplementuj bazę faktów dla systemu zarządzania magazynem. Baza danych powinna zawierać informacje o produktach, ich ilościach oraz lokalizacji w magazynie. Opracuj reguły, które umożliwią zarządzanie dostępnością produktów, ich przenoszeniem oraz uzupełnianiem stanu.    
 | SWI PROLOG

```python
% Deklaracja predykatów jako dynamicznych
:- dynamic produkt/3.

% Fakty o produktach (nazwa, ilość, sektor)
produkt(tv, 10, sektor_a).
produkt(laptop, 15, sektor_b).
produkt(glosnik, 20, sektor_a).
produkt(lodowka, 25, sektor_c).
produkt(telefon, 10, sektor_c).
produkt(drukarka, 20, sektor_b).

% Dostępność produktu
dostepny_produkt(Nazwa) :-
    produkt(Nazwa, Ilosc, _),
    Ilosc > 0, % Jeśli ilość jest większa od 0, produkt jest dostępny
    format('Produkt ~w jest dostępny.~n', [Nazwa]).

% Przenoszenie produktu
przenies_produkt(Nazwa, NowySektor) :-
    produkt(Nazwa, Ilosc, StarySektor),
    retract(produkt(Nazwa, Ilosc, StarySektor)), % Usuwa stary wpis
    assertz(produkt(Nazwa, Ilosc, NowySektor)), % Dodaje nowy wpis
    format('Produkt ~w został przeniesiony do ~w.~n', [Nazwa, NowySektor]).

% Uzupełnianie ilości produktu
uzupelnij_stan(Nazwa, DodajIlosc) :-
    produkt(Nazwa, StaraIlosc, Sektor),
    NowaIlosc is StaraIlosc + DodajIlosc, % Dodaje nową wartość do starej
    retract(produkt(Nazwa, StaraIlosc, Sektor)),
    assertz(produkt(Nazwa, NowaIlosc, Sektor)),
    format('Produkt ~w ma teraz ilość: ~w.~n', [Nazwa, NowaIlosc]).
```
