Stwórz system rekomendacji książek, który uwzględnia preferencje użytkowników, gatunki literackie oraz oceny książek. Twój system powinien być w stanie rekomendować książki danego gatunku użytkownikowi na podstawie jego preferencji oraz wcześniej przyznanych ocen.

```python
% Dynamiczne dodawanie i usuwanie ocen
:- dynamic ocena/3.

% Fakty o książkach (tytul, autor, gatunek)
ksiazka('Kamienie na szaniec', Kaminski, dramat).
ksiazka('Opowiesc wigilijna', Dickens, basn).
ksiazka('Percy Jackson', Riordan, fantasy).
ksiazka('Lalka', Prus, powiesc).
ksiazka('Hobbit', Tolkien, fantasy).
ksiazka('Wiedzmin', Sapkowski, fantasy).

% Oceny
ocena(uzytkownik1, 'Kamienie na szaniec', 9).
ocena(uzytkownik2, 'Lalka', 5).
ocena(uzytkownik2, 'Hobbit', 8).
ocena(uzytkownik2, 'Percy Jackson', 9).
ocena(uzytkownik1, 'Wiedzmin', 3).
ocena(uzytkownik1, 'Hobbit', 2).

% Preferencje
preferencje(uzytkownik1, dramat).
preferencje(uzytkownik2, fantasy).

% Regula polecane_ksiazki/2
polecane_ksiazki(X, Gatunek) :-
    preferencje(X, Gatunek), % czy użytkownik lubi dany gatunek
    findall(Tytul,
        (ksiazka(Tytul, _, Gatunek), ocena(X, Tytul, Ocena), Ocena >= 7), % Zbieramy książki z oceną >= 7
        ListaKsiazek),
    write('Polecane książki dla użytkownika '), write(X), write(' w gatunku '), write(Gatunek), write(': '), nl,
    write(ListaKsiazek), nl.

% Dodawanie oceny
dodaj_ocene(Uzytkownik, Tytul, Ocena) :-
    retractall(ocena(Uzytkownik, Tytul, _)), % Usuwamy istniejącą ocenę, jeśli jest
    assertz(ocena(Uzytkownik, Tytul, Ocena)). % Dodajemy nową ocenę

% Usuwanie oceny
usun_ocene(Uzytkownik, Tytul) :-
    retractall(ocena(Uzytkownik, Tytul, _)).

% np
% polecane_ksiazki(uzytkownik1, dramat).
% Polecane książki dla użytkownika uzytkownik1 w gatunku dramat: ['Kamienie na szaniec'].
```
