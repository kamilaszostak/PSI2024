Zaprojektuj i wdroż w Prologu bazę faktów dla systemu planowania trasy kuriera. Baza danych powinna zawierać informacje o miastach, połączeniach drogowych między nimi oraz czasie przejazdu. Opracuj reguły umożliwiające zapytania o dostępność dróg, obliczanie czasu przejazdu oraz wyznaczanie najkrótszej trasy.

```python
% Fakty o miastach
miasto(lublin).
miasto(gdansk).
miasto(poznan).
miasto(wroclaw).
miasto(rzeszow).

% Połączenia drogowe i czas przejazdu
droga(wroclaw, rzeszow, 2).
droga(lublin, gdansk, 5).
droga(lublin, poznan, 7).
droga(gdansk, poznan, 4).
droga(poznan, wroclaw, 3).

% Bezpośrednie połączenie między miastami
droga_bezposrednia(X, Y) :-
    droga(X, Y, _);
    droga(Y, X, _).

% Reguła określająca czas przejazdu przy połączeniu bezpośrednim
czas_przejazdu(X, Y, T) :-
    droga(X, Y, T);
    droga(Y, X, T).

% Reguła określająca czas przejazdu z połączeniami pośrednimi
czas_przejazdu(X, Y, T) :-
    czas_przejazdu(X, Y, T, []).

% Wyznaczanie najkrótszej trasy
najkrotsza_trasa(X, Y, Trasa, NajkrotszyCzas) :-
    znajdz_wszystkie_trasy(X, Y, Trasy),
    najkrotsza_z_listy(Trasy, Trasa, NajkrotszyCzas).

% Czas przejazdu z unikaniem odwiedzonych miast
czas_przejazdu(X, Y, T, Odwiedzone) :-
    droga(X, Z, T1),
    \+ member(Z, Odwiedzone),
    czas_przejazdu(Z, Y, T2, [X | Odwiedzone]),
    T is T1 + T2.
czas_przejazdu(X, Y, T, _) :-
    droga(X, Y, T).


```
