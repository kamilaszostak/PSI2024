Stwórz i zaimplementuj bazę faktów dla systemu rezerwacji lotów, który umożliwia zarządzanie lotniskami, połączeniami między nimi oraz cenami biletów. Opracuj reguły, które pozwolą na zapytania o bezpośrednie połączenia, koszt podróży oraz znalezienie najkrótszej trasy między lotniskami.  


```python
% Lista lotnisk
lotnisko(warszawa).
lotnisko(rzeszow).
lotnisko(krakow).
lotnisko(gdansk).
lotnisko(wroclaw).

% Informacje o lotach (lotnisko początkowe, lotnisko końcowe, koszt)
lot(warszawa, krakow, 200).
lot(rzeszow, gdansk, 80).
lot(warszawa, wroclaw, 150).
lot(krakow, rzeszow, 120).
lot(krakow, gdansk, 100).
lot(wroclaw, rzeszow, 50).
lot(warszawa, gdansk, 400).

% Sprawdzenie bezpośredniego połączenia
bezposredni_lot(X, Y, C) :-
    lot(X, Y, C).

%przyklad
bezposredni_lot(warszawa, krakow, C).
%odpowiedz
C= 200 true

% Obliczanie kosztu podróży między X a Y (bezpośrednio/przesiadkami)
koszt_lotu(X, Y, C) :-
    lot(X, Y, C). % bezpośredni lot
koszt_lotu(X, Y, C) :-
    lot(X, Z, C1),
    koszt_lotu(Z, Y, C2),
    C is C1 + C2. % oblicza sume

%przyklad
koszt_lotu(warszawa, gdansk, C).
%odpowiedz
C = 400 (bezposredni)
C= 300 (przez krakow i gdansk)

% Znajdowanie najkrotsza/najtanszej trasy
najtansza_trasa(X, Y, C, Trasa) :-
    najtansza_trasa(X, Y, C, [X], Trasa).

najtansza_trasa(X, Y, C, Odwiedzone, [X, Y]) :-
    lot(X, Y, C). % Bezpośrednia trasa
najtansza_trasa(X, Y, C, Odwiedzone, [X | Reszta]) :-
    lot(X, Z, C1),
    Z \= Y,
    \+ member(Z, Odwiedzone),
    najtansza_trasa(Z, Y, C2, [Z | Odwiedzone], Reszta),
    C is C1 + C2.
```
